```python
# Create series 
import numpy as np
import pandas as pd
x = [np.linspace(1,4,100,'')]
for i in  x:
    df =pd.DataFrame({'A': i})

y = [np.linspace(1,3,100,'')]
for i in  y:
    other =pd.DataFrame({'B': i})

z = [np.linspace(10000,30000,100,'')]
for i in  z:
    df2 =pd.DataFrame({'C': i})
print(df, other, df2)
```

```python
#Join columns
import numpy as np
import pandas as pd
df.join(other,lsuffix='',rsuffix='').join(df2 ,lsuffix='',rsuffix='')
```

```python
#rename
import pandas as pd
df.rename(columns={'A':'bedrs','B':'bathrs','C':'price_sqr-meter'})
```

```python


import matplotlib.pyplot as plt

#manipulated the data because I couldn't use .csv file to do this
#count done in excel using countif
labels = 'Assassinated','Died in captivity','Executed','Illness','Killed in Battle','Natural causes','Possibly assassinated','Suicide'
Count = [22,1,3,5,8,16,8,5]
explode = (0.1, 0, 0, 0, 0, 0, 0, 0)  # explodes 1st item because we are interested in that data
fig1, ax1 = plt.subplots()
ax1.pie(Count, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  #makes circle
plt.title("Roman Emperors that were assassinated")
plt.show()
```
